### 1.4 Outlier Detection

#### - Income Outliers (Z-Score > 3)

<b>
Z-Score > 3 = "To investigate records." It does NOT mean "Delete the records." Every flagged record must be reviewed individually.
<b>

```sql
-- Detect income outliers using Z-score
WITH IncomeStats AS (
    SELECT
        AVG(Income) AS AvgIncome,
        STDEV(Income) AS StdIncome
    FROM marketing_data
    WHERE Income IS NOT NULL
)
SELECT
    ID,
    Income,
    Year_Birth,
    Education,
    Marital_Status,
    ROUND((Income - s.AvgIncome) / s.StdIncome, 2) AS ZScore
FROM marketing_data m
CROSS JOIN IncomeStats s
WHERE ABS((Income - s.AvgIncome) / s.StdIncome) > 3
  AND Income IS NOT NULL
ORDER BY ZScore DESC;
```

<b>
Income in the ~ $153K–$162K range is 4 standard deviations above the mean, but still logically acceptable for senior professional income, so these rows were retained. Conversely, income values ​​exceeding $162K have no plausible business explanation and are likely typos, so these rows were removed from the dataset.
<b>

```sql
-- Delete rows with unrealistic income (above the reasonable limit of $162K)
DELETE FROM marketing_data
WHERE Income > 162397;
```

### 2.2 Marital Status Distribution

```sql
-- Marital status distribution with quality flag
SELECT
    Marital_Status,
    COUNT(*) AS CustomerCount,
    CAST(ROUND(100.0 * COUNT(*) / SUM(COUNT(*)) OVER(), 1) AS DECIMAL(5,2)) AS PctOfTotal,
    CASE
        WHEN Marital_Status IN ('YOLO','Absurd','Alone') THEN 'Non-Standard — Review'
        ELSE 'Standard'
    END AS DataQuality
FROM marketing_data
GROUP BY Marital_Status
ORDER BY CustomerCount DESC;
```

### 2.3 Total & Average Spend by Product Category

```sql
-- Product category spend analysis

WITH Totals AS (
    SELECT
        SUM(MntWines) AS TotalWines,
        SUM(MntFruits) AS TotalFruits,
        SUM(MntMeatProducts) AS TotalMeat,
        SUM(MntFishProducts) AS TotalFish,
        SUM(MntSweetProducts) AS TotalSweet,
        SUM(MntGoldProds) AS TotalGold,
        SUM(MntWines + MntFruits + MntMeatProducts +
            MntFishProducts + MntSweetProducts + MntGoldProds) AS GrandTotal
    FROM marketing_data
)
SELECT
    Category,
    TotalSpend,
    CAST(ROUND(TotalSpend * 100.0 / GrandTotal, 1) AS DECIMAL(10,2)) AS PctOfTotalSpend,
    CAST(ROUND(TotalSpend * 1.0 / 2240, 1) AS DECIMAL(10,2)) AS AvgSpendPerCustomer
FROM (
    SELECT 'Wines' AS Category, TotalWines AS TotalSpend, GrandTotal FROM Totals UNION ALL
    SELECT 'Meat', TotalMeat, GrandTotal FROM Totals UNION ALL
    SELECT 'Gold', TotalGold, GrandTotal FROM Totals UNION ALL
    SELECT 'Fish', TotalFish, GrandTotal FROM Totals UNION ALL
    SELECT 'Sweet', TotalSweet, GrandTotal FROM Totals UNION ALL
    SELECT 'Fruits', TotalFruits, GrandTotal FROM Totals
) sub
ORDER BY TotalSpend DESC;
```

<b>
Wines account for ~50% of all product spending, making them by far the most important revenue category. Meat products are second at ~28%. Together, Wines and Meat represent nearly 78% of all spend. Gold is the third-largest category at 7.3% (avg spend: $43.90 per customer). Fish, Sweets, and Fruits are relatively minor, each below 7%. Business implication: Wine promotions and premium meat bundles offer the highest revenue leverage. Gold products also present an upsell opportunity worth exploring. Fruits and Sweets may need targeted promotions to grow their share.
<b>

<b>

### 2.4 Purchase Volume by Channel

```sql
-- Total purchases by channel
SELECT
    'Store' AS Channel, SUM(NumStorePurchases) AS TotalPurchases FROM marketing_data UNION ALL
    SELECT 'Web', SUM(NumWebPurchases) FROM marketing_data UNION ALL
    SELECT 'Catalog', SUM(NumCatalogPurchases) FROM marketing_data UNION ALL
    SELECT 'Deals/Discounts', SUM(NumDealsPurchases) FROM marketing_data
ORDER BY TotalPurchases DESC;

-- Average per customer
SELECT
    CAST(ROUND(AVG(NumStorePurchases * 1.0), 2) AS DECIMAL(10,2)) AS AvgStorePurchases,
    CAST(ROUND(AVG(NumWebPurchases * 1.0), 2) AS DECIMAL(10,2)) AS AvgWebPurchases,
    CAST(ROUND(AVG(NumCatalogPurchases * 1.0), 2) AS DECIMAL(10,2)) AS AvgCatalogPurchases,
    CAST(ROUND(AVG(NumDealsPurchases * 1.0), 2) AS DECIMAL(10,2)) AS AvgDealsPurchases,
    CAST(ROUND(AVG(NumWebVisitsMonth * 1.0), 2) AS DECIMAL(10,2)) AS AvgWebVisitsPerMonth
FROM marketing_data;
```

<b>
In-store purchases are the dominant channel (~39% of all purchases), followed by Web at ~27%. Catalog purchases, while lower volume (~18%), represent a premium channel often associated with higher-value orders. Deals/Discounts account for ~16%, meaningful volume but at the cost of margins. Notably, avg web visits per month is ~5 but avg web purchases is ~4, suggesting reasonable web conversion but room to improve.
<b>

### 2.5 Campaign Acceptance Rates

```sql
-- Campaign performance comparison
SELECT
    Campaign,
    Accepted,
    TotalCustomers,
    CAST(ROUND(100.0 * Accepted / TotalCustomers, 2) AS DECIMAL(10,2)) AS AcceptanceRate_Pct
FROM (
    SELECT 'Campaign 1' AS Campaign, SUM(AcceptedCmp1) AS Accepted, COUNT(*) AS TotalCustomers FROM marketing_data UNION ALL
    SELECT 'Campaign 2', SUM(AcceptedCmp2), COUNT(*) FROM marketing_data UNION ALL
    SELECT 'Campaign 3', SUM(AcceptedCmp3), COUNT(*) FROM marketing_data UNION ALL
    SELECT 'Campaign 4', SUM(AcceptedCmp4), COUNT(*) FROM marketing_data UNION ALL
    SELECT 'Campaign 5', SUM(AcceptedCmp5), COUNT(*) FROM marketing_data UNION ALL
    SELECT 'Final (Response)', SUM(Response), COUNT(*) FROM marketing_data
) campaigns
ORDER BY AcceptanceRate_Pct DESC;
```

<b>
The most recent campaign (Response / Campaign 6) has the highest acceptance rate at 14.9%, suggesting improvement over time. Campaign 2 performed worst at 1.3% — far below the others. Campaigns 3, 4, and 5 are clustered around 7-7.4%, indicating consistent moderate performance. The upward trend culminating in Response suggests the business is learning from past campaigns and better targeting. Campaign 2 should be investigated for what made it fail (wrong segment, channel, or offer).

## 3. Segmentation Analysis

### 3.1 Income Bracket vs. Average Total Spend

```sql
-- Income-based spending analysis

WITH CustomerSpend AS (
    SELECT Income,
        MntWines+MntFruits+MntMeatProducts+
        MntFishProducts+MntSweetProducts+MntGoldProds AS TotalSpend,
        CASE
            WHEN Income < 20000 THEN '1. < $20K'
            WHEN Income BETWEEN 20000 AND 39999 THEN '2. $20K-$40K'
            WHEN Income BETWEEN 40000 AND 59999 THEN '3. $40K-$60K'
            WHEN Income BETWEEN 60000 AND 79999 THEN '4. $60K-$80K'
            WHEN Income BETWEEN 80000 AND 99999 THEN '5. $80K-$100K'
            WHEN Income >= 100000 THEN '6. $100K+'
        END AS IncomeBracket
    FROM marketing_data
    WHERE Income IS NOT NULL 
)
SELECT IncomeBracket, COUNT(*) AS Customers,
    ROUND(AVG(TotalSpend), 0) AS AvgTotalSpend,
    CAST(ROUND(AVG(Income), 0) AS DECIMAL(10,2)) AS AvgIncome,
    CAST(ROUND(AVG(TotalSpend * 1.0) / NULLIF(AVG(Income * 1.0), 0) * 100, 2) AS DECIMAL(10,2)) AS SpendPctOfIncome
FROM CustomerSpend
WHERE IncomeBracket IS NOT NULL
GROUP BY IncomeBracket
ORDER BY IncomeBracket;
```

<b>
There is a strong positive relationship between income and total spend up to the $80K–$100K bracket, where average spend peaks at $1,551, approximately 21x more than customers earning below $20K. However, the $100K+ group shows an unexpected drop in both absolute spend ($893) and spend-as-a-percentage-of-income (0.66%), though this bracket contains only 12 customers, making it statistically unreliable. The $80K–$100K bracket also shows the highest income allocation rate (1.82%), suggesting it represents the most engaged, premium-spending segment. Income remains a strong targeting variable, with the $80K–$100K bracket as the primary focus for premium campaigns.
<b>

<b>

### 3.2 Education vs. Spend & Campaign Response

```sql
-- Education level impact on spending and campaigns

WITH CustomerProfile AS (
    SELECT Education,
        MntWines+MntFruits+MntMeatProducts+
        MntFishProducts+MntSweetProducts+MntGoldProds AS TotalSpend,
        AcceptedCmp1+AcceptedCmp2+AcceptedCmp3+
        AcceptedCmp4+AcceptedCmp5+Response AS TotalCampaignsAccepted,
        Income
    FROM marketing_data
    WHERE Income IS NOT NULL 
)
SELECT Education, COUNT(*) AS Customers,
    CAST(ROUND(AVG(Income), 0) AS DECIMAL(10,2)) AS AvgIncome,
    CAST(ROUND(AVG(TotalSpend), 0) AS DECIMAL(10,2)) AS AvgTotalSpend,
    CAST(ROUND(AVG(TotalCampaignsAccepted * 1.0), 2) AS DECIMAL(10,2)) AS AvgCampaignsAccepted,
    CAST(ROUND(100.0 * SUM(CASE WHEN TotalCampaignsAccepted > 0 THEN 1 ELSE 0 END) / COUNT(*), 1) AS DECIMAL(10,2)) AS PctResponded_AnyCampaign
FROM CustomerProfile
GROUP BY Education
ORDER BY AvgTotalSpend DESC;
```

<b>
PhD customers lead across all metrics, highest average spend ($674), highest income ($56K), and highest campaign response rate (32.7%). Graduation-level customers are the largest segment (1,115 customers) and rank second in average spend ($622), outspending Master-level customers ($609), making them a high-volume, moderate-to-high value segment. Basic education customers spend the least ($81) and respond to campaigns at the lowest rate (13%). Strategy: Allocate premium campaign budget toward the PhD segment for quality, leverage Graduation-level customers for volume with value-focused messaging, and deprioritize Basic-education customers in campaign spend.
<b>

<b>

### 3.3 Children at Home vs. Spending & Channel

```sql
-- Impact of children on spending behavior

SELECT
    Kidhome + Teenhome AS TotalChildren,
    COUNT(*) AS Customers,
    CAST(ROUND(AVG(MntWines * 1.0), 0)          AS DECIMAL(10,2)) AS AvgWineSpend,
    CAST(ROUND(AVG(MntMeatProducts * 1.0), 0)   AS DECIMAL(10,2)) AS AvgMeatSpend,
    CAST(ROUND(AVG(MntFruits * 1.0), 0)         AS DECIMAL(10,2)) AS AvgFruitSpend,
    CAST(ROUND(AVG(MntSweetProducts * 1.0), 0)  AS DECIMAL(10,2)) AS AvgSweetSpend,

    CAST(ROUND(AVG(MntWines + MntFruits + MntMeatProducts +
        MntFishProducts + MntSweetProducts + MntGoldProds) * 1.0, 0) AS DECIMAL(10,2)) AS AvgTotalSpend,

    CAST(ROUND(AVG(NumWebPurchases * 1.0), 2)   AS DECIMAL(10,2)) AS AvgWebPurchases,
    CAST(ROUND(AVG(NumStorePurchases * 1.0), 2) AS DECIMAL(10,2)) AS AvgStorePurchases,
    CAST(ROUND(AVG(NumDealsPurchases * 1.0), 2) AS DECIMAL(10,2)) AS AvgDealsPurchases,

    CAST(ROUND(100.0 * AVG(CAST(Response AS FLOAT)), 1) AS DECIMAL(10,2)) AS LastCampaign_AcceptRate
FROM marketing_data
GROUP BY Kidhome + Teenhome
ORDER BY TotalChildren;
```

### 3.4 Recency Band vs. Spend & Purchases

```sql
-- Customer recency segmentation

WITH RecencyBands AS (
    SELECT *,
        CASE
            WHEN Recency <= 25 THEN '1. 0-25 days (Very Recent)'
            WHEN Recency <= 50 THEN '2. 26-50 days'
            WHEN Recency <= 75 THEN '3. 51-75 days'
            ELSE '4. 76+ days (Dormant)'
        END AS RecencyBand,

        MntWines + MntFruits + MntMeatProducts +
        MntFishProducts + MntSweetProducts + MntGoldProds AS TotalSpend,

        NumWebPurchases + NumCatalogPurchases + NumStorePurchases AS TotalPurchases

    FROM marketing_data
)
SELECT

    RecencyBand,
    COUNT(*) AS Customers,
    CAST(ROUND(AVG(TotalSpend), 0) AS DECIMAL(10,2)) AS AvgTotalSpend,
    CAST(ROUND(AVG(TotalPurchases * 1.0), 2) AS DECIMAL(10,2)) AS AvgPurchases,
    CAST(ROUND(100.0 * AVG(CAST(Response AS FLOAT)), 1) AS DECIMAL(10,2)) AS LastCampaign_AcceptRate

FROM RecencyBands
GROUP BY RecencyBand
ORDER BY RecencyBand;
```

<b>
Recency is a strong signal for campaign response rate, which clearly declines from 25.9% (0–25 days) to 7.1% (76+ days). However, spend and purchase frequency do not follow the same pattern, the 51–75 day band shows the highest spend ($650) and purchase frequency (12.90), suggesting these customers are high-value but less campaign-responsive. Dormant customers (76+ days) still spend $601 on average, indicating they have not disengaged from buying entirely. Strategy: Prioritize the 0–25 day segment for campaign activation (highest response rate), use re-engagement offers for the 51–75 day band to convert their high spend potential, and apply win-back messaging for 76+ day dormant customers.
<b>

### 3.5 Web Visits vs. Web Purchases

```sql
-- Web behavior analysis

SELECT
    NumWebVisitsMonth AS WebVisits,
    COUNT(*) AS Customers,
    CASE WHEN NumWebVisitsMonth = 0 THEN 0 ELSE ROUND(AVG(NumWebPurchases * 1.0), 2) END AS AvgWebPurchases,

    CASE WHEN NumWebVisitsMonth = 0 THEN 0 ELSE ROUND(AVG(NumWebPurchases * 1.0) / NULLIF(NumWebVisitsMonth, 0), 3) END AS PurchasePerVisit,

    CASE WHEN NumWebVisitsMonth = 0 THEN 0 ELSE ROUND(AVG(MntWines + MntFruits + MntMeatProducts +
        MntFishProducts + MntSweetProducts + MntGoldProds) * 1.0, 0) END AS AvgTotalSpend

FROM marketing_data
GROUP BY NumWebVisitsMonth
ORDER BY NumWebVisitsMonth;
```

<b>
Interestingly, customers who visit the website very frequently (6-8+ times/month) do NOT necessarily make more web purchases, and their average total spend drops sharply (from $1,238 at 1 visit to $297 at 7 visits). The PurchasePerVisit ratio declines from 3.65 to below 0.5, suggesting heavy browsers are comparing prices or abandoning carts without converting. Notably, customers with 0 web visits still spend an average of $1,031, indicating a strong offline/catalog buyer segment. This is a signal for A/B testing personalized offers for heavy browsers and exit-intent campaigns.

## 4. Customer Lifecycle Analysis

### 4.1 New Customer Enrollment by Month/Year

```sql
-- Customer acquisition trends

SELECT
    YEAR(Dt_Customer) AS EnrollYear,
    MONTH(Dt_Customer) AS EnrollMonth,
    DATENAME(MONTH, Dt_Customer) AS MonthName,
    COUNT(*) AS NewCustomers,

    SUM(COUNT(*)) OVER (ORDER BY YEAR(Dt_Customer), MONTH(Dt_Customer)
                        ROWS UNBOUNDED PRECEDING) AS CumulativeCustomers
                        
FROM marketing_data
GROUP BY
    YEAR(Dt_Customer),
    MONTH(Dt_Customer),
    DATENAME(MONTH, Dt_Customer)
ORDER BY
    EnrollYear,
    EnrollMonth;
```

<b>
The enrollment data spans July 2012–June 2014, totaling 2,236 customers. July 2012 shows only 2 enrollments, likely representing the business launch or first month of data. 2013 was the most consistent acquisition year (88–117 customers/month). Enrollment toward the end of the period shows notable volatility, peaking at 111 in May 2014 but dropping to 74 in June 2014, the lowest month in the dataset, suggesting the acquisition funnel warrants attention. Months with high spikes (Aug 2012: 118, Oct 2013: 117) may align with promotional events. The window function (SUM OVER) provides a running total without requiring a self-join.
<b>

<b>

### 4.2 Customer Tenure vs. Total Lifetime Spend

```sql
-- Tenure-based value analysis

WITH Tenure AS (
    SELECT
        ID,
        DATEDIFF(DAY, Dt_Customer, '2015-01-01') AS TenureDays,
        MntWines + MntFruits + MntMeatProducts +
        MntFishProducts + MntSweetProducts + MntGoldProds AS TotalSpend,
        CASE
            WHEN DATEDIFF(DAY, Dt_Customer, '2015-01-01') < 365 THEN '< 1 Year'
            WHEN DATEDIFF(DAY, Dt_Customer, '2015-01-01') < 730 THEN '1-2 Years'
            WHEN DATEDIFF(DAY, Dt_Customer, '2015-01-01') < 1095 THEN '2-3 Years'
            ELSE '3+ Years'
        END AS TenureBand
    FROM marketing_data
)
SELECT
    TenureBand,
    COUNT(*) AS Customers,
    ROUND(AVG(TenureDays), 0) AS AvgTenureDays,
    ROUND(AVG(TotalSpend), 0) AS AvgLifetimeSpend,
    ROUND(MIN(TotalSpend), 0) AS MinSpend,
    ROUND(MAX(TotalSpend), 0) AS MaxSpend
FROM Tenure
GROUP BY TenureBand
ORDER BY AvgTenureDays;
```

<b>
Consistent upward trend: <1 Year ($492) → 1-2 Years ($602) → 2-3 Years ($741). Customers with 2-3 years tenure spend 51% more per customer than new customers, a strong justification for loyalty program investment. While 2-3 Years has the highest average spend per customer ($741), the 1-2 Years cohort generates the most total revenue (~$713K) due to its larger size (1,185 customers). Both segments are priority targets, the former for per-customer value, the latter for volume.
<b>

## 5. Additional Analyses

### 5.1  Country-Level Performance

```sql
SELECT
    Country,
    COUNT(*) AS Customers,
    CAST(ROUND(AVG(Income), 0) AS INTEGER)  AS AvgIncome,

    ROUND(AVG(MntWines + MntFruits + MntMeatProducts +
        MntFishProducts + MntSweetProducts + MntGoldProds) * 1.0, 0) AS AvgTotalSpend,

    CAST(ROUND(AVG(NumWebPurchases * 1.0), 2) AS REAL)     AS AvgWebPurchases,
    CAST(ROUND(AVG(NumStorePurchases * 1.0), 2) AS REAL)   AS AvgStorePurchases,
    ROUND(100.0 * AVG(CAST(Response AS FLOAT)), 1)         AS LastCampaign_AcceptRate,
    ROUND(100.0 * AVG(CAST(Complain AS FLOAT)), 2)         AS ComplaintRate_Pct,

    RANK() OVER (ORDER BY AVG(MntWines+MntFruits+MntMeatProducts+
        MntFishProducts+MntSweetProducts+MntGoldProds) DESC) AS SpendRank

FROM marketing_data
WHERE Income IS NOT NULL
GROUP BY Country
HAVING COUNT(*) >= 10  -- Exclude micro-markets (Mexico: 3 customers)
ORDER BY AvgTotalSpend DESC;
```

Among this top tier, Saudi Arabia has the highest campaign acceptance rate at 15.5%, making it the most responsive high-value market.

Spain, ranking 5th in average spend ($602), deserves special attention: it is the largest market by far with 1,092 customers — nearly half the dataset — and holds the highest campaign acceptance rate overall at 16.1%. This combination of scale and responsiveness makes Spain uniquely powerful for broad campaigns.

At the lower end, Australia ($582) and India ($532) show the lowest average spend, suggesting lower purchasing power or a different product-market fit. USA and Australia both show 0% complaint rate — USA alongside a top-tier spend of $631, a strong signal of customer satisfaction.

### 5.2  Multi-Campaign Responders — VIP Loyalist Profile

```sql
-- 5.2 Multi-Campaign Responder Profile 

WITH CampaignEngagement AS (
    SELECT
        ID,
        CAST(2015 - Year_Birth AS INT)                    AS Age,

        Education,
        Marital_Status,
        Income,
        Kidhome + Teenhome                                AS TotalChildren,

        AcceptedCmp1 + AcceptedCmp2 + AcceptedCmp3 +
        AcceptedCmp4 + AcceptedCmp5 + Response            AS TotalCampaignsAccepted,

        MntWines + MntFruits + MntMeatProducts +
        MntFishProducts + MntSweetProducts + MntGoldProds AS TotalSpend,

        NumWebPurchases + NumCatalogPurchases + 
        NumStorePurchases                                 AS TotalPurchases

    FROM marketing_data
    WHERE Income IS NOT NULL 
      
)

SELECT

    TotalCampaignsAccepted          AS CampaignsAccepted,
    COUNT(*)                        AS Customers,
    CAST(AVG(Age * 1.0) AS INT)     AS AvgAge,
    CAST(ROUND(AVG(Income), 0)      AS INTEGER)  AS AvgIncome,
    ROUND(AVG(TotalSpend), 0)       AS AvgTotalSpend,
    CAST(AVG(TotalPurchases * 1.0)  AS DECIMAL(5,1)) AS AvgPurchases,
    CAST(AVG(TotalChildren * 1.0)   AS DECIMAL(5,1)) AS AvgChildren

FROM CampaignEngagement
GROUP BY TotalCampaignsAccepted
ORDER BY TotalCampaignsAccepted DESC;
```

Age is not a distinguishing factor, the 4-campaign group (avg 43) is actually slightly younger than the general population. This profile is the ideal lookalike audience for paid acquisition campaigns. These customers should receive a dedicated loyalty track: early access to new products, exclusive discounts, and personal account management.

### BQ2: Campaign Success — Rate + Responder Quality

```sql
-- Detailed campaign performance metrics

SELECT 'Campaign 1' AS Campaign, SUM(AcceptedCmp1) AS Accepted,
    ROUND(100.0*AVG(CAST(AcceptedCmp1 AS FLOAT)),1) AS AcceptRate_Pct,

    ROUND(AVG(CASE WHEN AcceptedCmp1=1 THEN
        MntWines+MntFruits+MntMeatProducts+MntFishProducts+MntSweetProducts+MntGoldProds END),0)
        AS AvgSpend_Responders,

    ROUND(AVG(CASE WHEN AcceptedCmp1=0 THEN
        MntWines+MntFruits+MntMeatProducts+MntFishProducts+MntSweetProducts+MntGoldProds END),0)
        AS AvgSpend_NonResponders

FROM marketing_data

UNION ALL

SELECT 'Campaign 2', SUM(AcceptedCmp2), ROUND(100.0*AVG(CAST(AcceptedCmp2 AS FLOAT)),1),
    ROUND(AVG(CASE WHEN AcceptedCmp2=1 THEN MntWines+MntFruits+MntMeatProducts+MntFishProducts+MntSweetProducts+MntGoldProds END),0),
    ROUND(AVG(CASE WHEN AcceptedCmp2=0 THEN MntWines+MntFruits+MntMeatProducts+MntFishProducts+MntSweetProducts+MntGoldProds END),0)
FROM marketing_data

UNION ALL

SELECT 'Campaign 3', SUM(AcceptedCmp3), ROUND(100.0*AVG(CAST(AcceptedCmp3 AS FLOAT)),1),
    ROUND(AVG(CASE WHEN AcceptedCmp3=1 THEN MntWines+MntFruits+MntMeatProducts+MntFishProducts+MntSweetProducts+MntGoldProds END),0),
    ROUND(AVG(CASE WHEN AcceptedCmp3=0 THEN MntWines+MntFruits+MntMeatProducts+MntFishProducts+MntSweetProducts+MntGoldProds END),0)
FROM marketing_data

UNION ALL

SELECT 'Campaign 4', SUM(AcceptedCmp4), ROUND(100.0*AVG(CAST(AcceptedCmp4 AS FLOAT)),1),
    ROUND(AVG(CASE WHEN AcceptedCmp4=1 THEN MntWines+MntFruits+MntMeatProducts+MntFishProducts+MntSweetProducts+MntGoldProds END),0),
    ROUND(AVG(CASE WHEN AcceptedCmp4=0 THEN MntWines+MntFruits+MntMeatProducts+MntFishProducts+MntSweetProducts+MntGoldProds END),0)
FROM marketing_data

UNION ALL

SELECT 'Campaign 5', SUM(AcceptedCmp5), ROUND(100.0*AVG(CAST(AcceptedCmp5 AS FLOAT)),1),
    ROUND(AVG(CASE WHEN AcceptedCmp5=1 THEN MntWines+MntFruits+MntMeatProducts+MntFishProducts+MntSweetProducts+MntGoldProds END),0),
    ROUND(AVG(CASE WHEN AcceptedCmp5=0 THEN MntWines+MntFruits+MntMeatProducts+MntFishProducts+MntSweetProducts+MntGoldProds END),0)
FROM marketing_data

UNION ALL

SELECT 'Final (Response)', SUM(Response), ROUND(100.0*AVG(CAST(Response AS FLOAT)),1),
    ROUND(AVG(CASE WHEN Response=1 THEN MntWines+MntFruits+MntMeatProducts+MntFishProducts+MntSweetProducts+MntGoldProds END),0),
    ROUND(AVG(CASE WHEN Response=0 THEN MntWines+MntFruits+MntMeatProducts+MntFishProducts+MntSweetProducts+MntGoldProds END),0)
FROM marketing_data

ORDER BY AcceptRate_Pct DESC;
```

<b>
The final campaign (Response) leads decisively on acceptance rate (14.9%), more than double any individual campaign, but ranks 5th out of 6 in responder spend quality ($987). The real quality winner is Campaign 5, whose responders average $1,614 in spend with the highest lift ratio (3.06x vs non-responders). Campaign 1 follows closely at $1,482 (2.72x lift). Campaign 2, while the lowest in acceptance rate (1.3%), actually attracts high-quality responders ($1,307), its issue is low reach, not low quality. Campaign 3 shows the weakest responder lift (1.21x), suggesting poor targeting precision. Strategic recommendation: the final campaign's targeting model should be retained for volume, while Campaign 5's targeting criteria should be reverse-engineered for premium segment campaigns.

### BQ4: Product Performance — Revenue & Penetration

```sql
-- Product category performance metrics
WITH ProductMetrics AS (
    SELECT
        SUM(MntWines)              AS TotalWines,
        SUM(MntFruits)             AS TotalFruits,
        SUM(MntMeatProducts)       AS TotalMeat,
        SUM(MntFishProducts)       AS TotalFish,
        SUM(MntSweetProducts)      AS TotalSweet,
        SUM(MntGoldProds)          AS TotalGold,

        COUNT(*)                   AS N,

        SUM(CASE WHEN MntWines > 0 THEN 1 ELSE 0 END)          AS BuyersWine,
        SUM(CASE WHEN MntFruits > 0 THEN 1 ELSE 0 END)         AS BuyersFruit,
        SUM(CASE WHEN MntMeatProducts > 0 THEN 1 ELSE 0 END)   AS BuyersMeat,
        SUM(CASE WHEN MntFishProducts > 0 THEN 1 ELSE 0 END)   AS BuyersFish,
        SUM(CASE WHEN MntSweetProducts > 0 THEN 1 ELSE 0 END)  AS BuyersSweet,
        SUM(CASE WHEN MntGoldProds > 0 THEN 1 ELSE 0 END)      AS BuyersGold
    FROM marketing_data
)
SELECT Product, TotalRevenue, PctOfRevenue, BuyerPenetration_Pct
FROM (
    SELECT 'Wines' AS Product,
           TotalWines AS TotalRevenue,
           CAST(ROUND(100.0*TotalWines /(TotalWines+TotalFruits+TotalMeat+TotalFish+TotalSweet+TotalGold),1) AS FLOAT) AS PctOfRevenue,
           CAST(ROUND(100.0*BuyersWine/N, 1) AS FLOAT) AS BuyerPenetration_Pct
    FROM ProductMetrics

    UNION ALL

    SELECT 'Meat', TotalMeat,
           ROUND(100.0*TotalMeat/(TotalWines+TotalFruits+TotalMeat+TotalFish+TotalSweet+TotalGold),1),
           ROUND(100.0*BuyersMeat/N,1)
    FROM ProductMetrics

    UNION ALL

    SELECT 'Gold', TotalGold,
           ROUND(100.0*TotalGold/(TotalWines+TotalFruits+TotalMeat+TotalFish+TotalSweet+TotalGold),1),
           ROUND(100.0*BuyersGold/N,1)
    FROM ProductMetrics

    UNION ALL

    SELECT 'Fish', TotalFish,
           ROUND(100.0*TotalFish/(TotalWines+TotalFruits+TotalMeat+TotalFish+TotalSweet+TotalGold),1),
           ROUND(100.0*BuyersFish/N,1)
    FROM ProductMetrics

    UNION ALL

    SELECT 'Sweet', TotalSweet,
           ROUND(100.0*TotalSweet/(TotalWines+TotalFruits+TotalMeat+TotalFish+TotalSweet+TotalGold),1),
           ROUND(100.0*BuyersSweet/N,1)
    FROM ProductMetrics

    UNION ALL

    SELECT 'Fruits', TotalFruits,
           ROUND(100.0*TotalFruits/(TotalWines+TotalFruits+TotalMeat+TotalFish+TotalSweet+TotalGold),1),
           ROUND(100.0*BuyersFruit/N,1)
    FROM ProductMetrics
) products

ORDER BY TotalRevenue DESC;
```

<b>
Wines dominate revenue with ~50% share, while Meat leads buyer penetration at 100%, every customer buys meat. Both products are true category anchors. Gold products show high penetration (97.3%) relative to their revenue share (7.3%), suggesting customers buy frequently but in small amounts, a clear upsell opportunity. Fish, Fruits, and Sweets occupy the bottom in both revenue and penetration (81–83%), though even these "niche" products still reach over 80% of customers. Bundling strategies (e.g., "Wine + Chocolate" or "Meat + Fish" gift sets) may help increase basket size for these lower-revenue categories rather than relying on standalone promotion.

### BQ5: Channel Performance Analysis

```sql
-- Sales channel performance metrics

WITH ChannelStats AS (
    SELECT
        SUM(NumWebPurchases)        AS TotalWebPurch,
        SUM(NumCatalogPurchases)    AS TotalCatalogPurch,
        SUM(NumStorePurchases)      AS TotalStorePurch,
        SUM(NumDealsPurchases)      AS TotalDealsPurch,

        SUM(MntWines+MntFruits+MntMeatProducts+
            MntFishProducts+MntSweetProducts+MntGoldProds) AS TotalSpend,

        SUM(NumWebPurchases+NumCatalogPurchases+NumStorePurchases) AS TotalPurch,
        
        COUNT(*) AS N

    FROM marketing_data
)
SELECT
    Channel,
    PurchaseVolume,
    CAST(ROUND(100.0 * PurchaseVolume / (TotalWebPurch+TotalCatalogPurch+TotalStorePurch+TotalDealsPurch), 1) AS FLOAT) AS VolShare_Pct,
    CAST(ROUND(TotalSpend * 1.0 / NULLIF(PurchaseVolume, 0), 2) AS FLOAT) AS RevenuePerPurchase_Proxy
FROM (
    SELECT 'Web' AS Channel, TotalWebPurch AS PurchaseVolume,
           TotalSpend, TotalWebPurch, TotalCatalogPurch, TotalStorePurch, TotalDealsPurch FROM ChannelStats
    UNION ALL
    SELECT 'Catalog', TotalCatalogPurch, TotalSpend,
           TotalWebPurch, TotalCatalogPurch, TotalStorePurch, TotalDealsPurch FROM ChannelStats
    UNION ALL
    SELECT 'Store', TotalStorePurch, TotalSpend,
           TotalWebPurch, TotalCatalogPurch, TotalStorePurch, TotalDealsPurch FROM ChannelStats
    UNION ALL
    SELECT 'Deals', TotalDealsPurch, TotalSpend,
           TotalWebPurch, TotalCatalogPurch, TotalStorePurch, TotalDealsPurch FROM ChannelStats
) sub
ORDER BY PurchaseVolume DESC;
```

<b>
Deals is actually the most revenue-efficient channel at $260.52 per purchase, suggesting customers tend to buy in larger quantities when promotions are available. Rather than eroding margins, deals appear to drive higher-value transactions. The real underperformer in efficiency is Store ($104.56 per purchase) despite leading in volume (39% share).
