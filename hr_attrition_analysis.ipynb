## 1. Data Exploration


### 1.1 Data Overview

```python
---- ROW COUNT
SELECT COUNT(*) AS TotalRows 
FROM dbo.HR_Employee_Attrition
```

```python
---- COLUMN TYPES
SELECT 
    COLUMN_NAME,
    DATA_TYPE,
    CHARACTER_MAXIMUM_LENGTH  AS Max_Length,
    IS_NULLABLE
FROM INFORMATION_SCHEMA.COLUMNS
WHERE TABLE_NAME = 'HR_Employee_Attrition'
ORDER BY ORDINAL_POSITION
```

```python
---- NULL CHECK 
SELECT
    SUM(CASE WHEN Age                      IS NULL THEN 1 ELSE 0 END) AS Age,
    SUM(CASE WHEN Attrition                IS NULL THEN 1 ELSE 0 END) AS Attrition,
    SUM(CASE WHEN BusinessTravel           IS NULL THEN 1 ELSE 0 END) AS BusinessTravel,
    SUM(CASE WHEN DailyRate                IS NULL THEN 1 ELSE 0 END) AS DailyRate,
    SUM(CASE WHEN Department               IS NULL THEN 1 ELSE 0 END) AS Department,
    SUM(CASE WHEN DistanceFromHome         IS NULL THEN 1 ELSE 0 END) AS DistanceFromHome,
    SUM(CASE WHEN Education                IS NULL THEN 1 ELSE 0 END) AS Education,
    SUM(CASE WHEN EducationField           IS NULL THEN 1 ELSE 0 END) AS EducationField,
    SUM(CASE WHEN EmployeeCount            IS NULL THEN 1 ELSE 0 END) AS EmployeeCount,
    SUM(CASE WHEN EmployeeNumber           IS NULL THEN 1 ELSE 0 END) AS EmployeeNumber,
    SUM(CASE WHEN EnvironmentSatisfaction  IS NULL THEN 1 ELSE 0 END) AS EnvironmentSatisfaction,
    SUM(CASE WHEN Gender                   IS NULL THEN 1 ELSE 0 END) AS Gender,
    SUM(CASE WHEN HourlyRate               IS NULL THEN 1 ELSE 0 END) AS HourlyRate,
    SUM(CASE WHEN JobInvolvement           IS NULL THEN 1 ELSE 0 END) AS JobInvolvement,
    SUM(CASE WHEN JobLevel                 IS NULL THEN 1 ELSE 0 END) AS JobLevel,
    SUM(CASE WHEN JobRole                  IS NULL THEN 1 ELSE 0 END) AS JobRole,
    SUM(CASE WHEN JobSatisfaction          IS NULL THEN 1 ELSE 0 END) AS JobSatisfaction,
    SUM(CASE WHEN MaritalStatus            IS NULL THEN 1 ELSE 0 END) AS MaritalStatus,
    SUM(CASE WHEN MonthlyIncome            IS NULL THEN 1 ELSE 0 END) AS MonthlyIncome,
    SUM(CASE WHEN MonthlyRate              IS NULL THEN 1 ELSE 0 END) AS MonthlyRate,
    SUM(CASE WHEN NumCompaniesWorked       IS NULL THEN 1 ELSE 0 END) AS NumCompaniesWorked,
    SUM(CASE WHEN Over18                   IS NULL THEN 1 ELSE 0 END) AS Over18,
    SUM(CASE WHEN OverTime                 IS NULL THEN 1 ELSE 0 END) AS OverTime,
    SUM(CASE WHEN PercentSalaryHike        IS NULL THEN 1 ELSE 0 END) AS PercentSalaryHike,
    SUM(CASE WHEN PerformanceRating        IS NULL THEN 1 ELSE 0 END) AS PerformanceRating,
    SUM(CASE WHEN RelationshipSatisfaction IS NULL THEN 1 ELSE 0 END) AS RelationshipSatisfaction,
    SUM(CASE WHEN StandardHours            IS NULL THEN 1 ELSE 0 END) AS StandardHours,
    SUM(CASE WHEN StockOptionLevel         IS NULL THEN 1 ELSE 0 END) AS StockOptionLevel,
    SUM(CASE WHEN TotalWorkingYears        IS NULL THEN 1 ELSE 0 END) AS TotalWorkingYears,
    SUM(CASE WHEN TrainingTimesLastYear    IS NULL THEN 1 ELSE 0 END) AS TrainingTimesLastYear,
    SUM(CASE WHEN WorkLifeBalance          IS NULL THEN 1 ELSE 0 END) AS WorkLifeBalance,
    SUM(CASE WHEN YearsAtCompany           IS NULL THEN 1 ELSE 0 END) AS YearsAtCompany,
    SUM(CASE WHEN YearsInCurrentRole       IS NULL THEN 1 ELSE 0 END) AS YearsInCurrentRole,
    SUM(CASE WHEN YearsSinceLastPromotion  IS NULL THEN 1 ELSE 0 END) AS YearsSinceLastPromotion,
    SUM(CASE WHEN YearsWithCurrManager     IS NULL THEN 1 ELSE 0 END) AS YearsWithCurrManager
FROM dbo.HR_Employee_Attrition;
```

### 1.2 Attrition by Demographics

```python
-- Attrition Distribution
SELECT
    Attrition,
    COUNT(*)                                      AS Total,
    CAST(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER() AS DECIMAL(5,2)) AS Percentage
FROM dbo.HR_Employee_Attrition
GROUP BY Attrition;
```

### 1.3 Attrition by Department

```python
---- By Department
SELECT
    Department,
    COUNT(*)                                                        AS Total,
    CAST(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER() AS DECIMAL(5,2))   AS Percentage
FROM dbo.HR_Employee_Attrition
GROUP BY Department
ORDER BY Total DESC;
```

```python
---- By Job Role
SELECT
    JobRole,
    COUNT(*)                                                          AS Total,
    CAST(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER() AS DECIMAL(5,2))   AS Percentage
FROM dbo.HR_Employee_Attrition
GROUP BY JobRole
ORDER BY Total DESC;
```

```python

---- By Education Field
SELECT
    EducationField,
    COUNT(*)                                                          AS Total,
    CAST(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER() AS DECIMAL(5,2))   AS Percentage
FROM dbo.HR_Employee_Attrition
GROUP BY EducationField
ORDER BY Total DESC;
```
